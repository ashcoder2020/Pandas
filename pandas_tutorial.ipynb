## Create series using dictionary

```python

dict_s = pd.Series({'a':1,'b':2})       #create series using dictiopnary
print(dict_s)

```

```python

s4 = pd.Series([1,2,3,4,5])
print(s4)                          #access element by index
print(s4[4])

print(s4[0:3])                          #slicing
print(max(s4))                          #max value in series
print(min(s4))
print(s4>3)                             #gives value greate than 3

print(s4)
```

# Handling Missing Values

```python
df = pd.read_csv('sheet.csv')   
```

### keep null values same as it is not replace with NaN

```python
df = pd.read_csv('sheet.csv',keep_default_na=False)                 #keep values same if blank then blank not replace by Nan  
df
```

```python
df = pd.read_csv('sheet.csv',na_filter=False)               #boost performance when no null values using na_filter      
df
```

### Check null values in file

```python
df.isnull().sum()             #give the total number of null values in table column wise         
```

```python
df.isnull().sum().sum()                 #gives total null values in the table
```

```python
df.notnull()
```

```python
df.notnull().sum().sum()
```

### drop the null values

```python
df = pd.read_csv('sheet.csv')
df
```

```python
df.dropna()       #drop null value row
```

```python
df.dropna(axis = 1)                        #0-row, 1-clm   #drop null value column
```

```python
df.dropna(thresh=2)                          #maximum 2 null values available if more than two null in single row then drop this row
```

### fill the null values

```python
df = pd.read_csv('sheet.csv')
df
```

```python
df.fillna(0)                #fill null value to zero
```

```python
df.fillna(2)              #fill null value to 2 which is written in fillna()
```

```python
df.fillna({'Student id' : 'none', 'mobile game': 0})                  #fill null value with specified column with specific value
```

```python
df.fillna(method='ffill')         #bfill takes backword value of clm       #previous value fill to null value place
```

```python
df.fillna(axis=0,method='ffill')    #fill forward value from row 
```

```python
df.fillna(0,limit=1)            #specific nan value filled using these(only one missing value fill)
```

## replacing values

value to dataframe are replaced with other values dynamically
syntax: DataFrame.replace()

parameters: to_replace and limit

```python
df = pd.read_csv('sheet.csv')
df.replace(to_replace = 1001,value='ashu') 
```

```python
df.replace(0,method = 'bfill',limit=1)               #change only one value because of limit
```

## loc() and iloc() for values operation

access a group of rows and clms by label(s) or a boolean aray


```python
df = pd.read_csv('sheet.csv')
df.loc[0]                 #gives 0th row values
```

```python
df.loc[4]                             #gives nth  row
```

```python
df.loc[[0,1,2]]                          #gives given index calues with all values
```

```python
df.loc[4,'percentage']                   #gives ith index and given clm value
```

```python
df.loc[0:3,'percentage']                 #gives slicing and clm name which output indexgiven row in clm name
```

```python
df.iloc[0]                                   #return zero index value
```

```python
df.iloc[:,0]                         #(row,clm) cant do label indexing 
```

```python
df.iloc[[0,1]]                       #print given rows
```
